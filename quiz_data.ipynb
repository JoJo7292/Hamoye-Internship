# MULTIPLE LINEAR REGRESSION

```python
norm_df.head()
```

```python
features_df = energy_df.drop(columns=['Appliances'])
target_ = energy_df['Appliances']
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_df, target_, test_size=0.3, random_state=42)
lin_mod = LinearRegression()
lin_mod.fit(x_train, y_train)
y_values = lin_mod.predict(x_test)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
#comparing the effects pf regularization
def get_weights_df(lin_mod, feat, col_name):
    #this function returns the weight of every feature
    weights = pd.Series(lin_mod.coef_,feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
lin_mod_weights = get_weights_df(lin_mod, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_Weight')
```

```python
final_weights = pd.merge(lin_mod_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on="Features")
```
